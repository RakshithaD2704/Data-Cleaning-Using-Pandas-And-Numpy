```python
import numpy as np
import pandas as pd
```

```python
data.columns = data.columns.str.lower().str.replace(' ','_')
```

```python
data.columns
```

```python
data.head()
```

```python
avg_age = data.age.mean()
```

```python
data['age'] = data['age'].fillna(avg_age)
data['age'] = data['age'].round(decimals=1)
data['age']
```

```python
data['age'].isna().sum()
```

## Salary

```python
data['salary'] = data['salary'].apply(lambda x:x.replace('$',''))
```

```python
data.head()
```

```python
data['salary']
```

```python
data['age'] = data['age'].astype(int)
```

## cleaning established column

```python
data["established"] = data["established"].replace(-1,np.nan)
data["established"].fillna('unknown', inplace=True)
```

## Location column

```python
data['location']
```

```python
data['location'] = data['location'].str.replace("Australia Aus", "Australia,Aus")
data['location_city'] = data['location'].str.split(",", expand=True).get(0)
data['city_sign'] = data['location'].str.split(",", expand=True).get(1)
data.drop('location', inplace=True, axis=1)
```

```python
data.head()
```

## Rating column

```python
data['rating'] = data["rating"].replace(-1,np.nan)

data['rating'] = data["rating"].fillna(data['rating'].mean()).round(decimals=1)
```

## Easy apply

```python
data['easy_apply'] = data['easy_apply'].replace("-1", "True")
```

## Salary column

```python
data['salary'].iloc[0]

data['salary'] = data['salary'].apply(lambda x:x.replace("$",''))
data['salary'] = data['salary'].apply(lambda x:x.replace("k", ''))
```
